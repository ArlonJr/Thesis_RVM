```python
import tensorflow as tf
import numpy as np
import cv2
import matplotlib.pyplot as plt
from keras.applications.mobilenet_v2 import preprocess_input
import os

def predict_single_image(model, image_path, categories, target_size=(224, 224)):
    """
    Predict the class of a single image.
    
    Parameters:
    model: The trained model
    image_path: Path to the image file
    categories: List of category names [category_a, category_b]
    target_size: Image size expected by the model
    
    Returns:
    predicted_class: The predicted class index (0 or 1)
    predicted_label: The predicted class name
    confidence: Confidence score (0 to 1)
    """
    try:
        # Read the image
        img = cv2.imread(image_path)
        if img is None:
            print(f"Error: Could not read image {image_path}")
            return None, None, None
            
        # Convert BGR to RGB
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        
        # Resize image
        img = cv2.resize(img, target_size).astype(np.float32)
        
        # Preprocess for MobileNetV2
        img = preprocess_input(img)
        
        # Add batch dimension
        img_batch = np.expand_dims(img, axis=0)
        
        # Make prediction
        prediction = model.predict(img_batch)[0][0]
        
        # Convert to class
        predicted_class = 1 if prediction > 0.5 else 0
        predicted_label = categories[predicted_class]
        
        # Get confidence
        confidence = prediction if predicted_class == 1 else 1 - prediction
        
        if(predicted_class == 0 and confidence < 0.8):
            predicted_label = categories[1]
        
        print(f"\nResults for {os.path.basename(image_path)}:")
        print(f"Prediction: {predicted_label} (Class {predicted_class})")
        print(f"Confidence: {confidence:.4f}")
        
        # Optional: visualize the image with its prediction
        plt.figure(figsize=(6, 6))
        
        # Convert image back for visualization
        display_img = cv2.resize(cv2.imread(image_path), target_size)
        display_img = cv2.cvtColor(display_img, cv2.COLOR_BGR2RGB)
        
        plt.imshow(display_img)
        plt.title(f"Predicted: {predicted_label}\nConfidence: {confidence:.4f}")
        plt.axis('off')
        plt.show()
        
        return predicted_class, predicted_label, confidence
        
    except Exception as e:
        print(f"Error processing image {image_path}: {e}")
        return None, None, None

# Direct input parameters for notebook environment
model_path = "C:/Users/Jarl Madamba/Documents/RVM_Binary_ML/Trained Models/binary4/RVM_Model_mk6.keras"
image_path = "C:/Users/Jarl Madamba/Downloads/royal_2.jpg"
categories = ["plastic", "waste"]  # Replace with your actual category names

# Load the model
print(f"Loading model from {model_path}...")
try:
    model = tf.keras.models.load_model(model_path)
    print("Model loaded successfully!")
except Exception as e:
    print(f"Error loading model: {e}")

# Make prediction
print(f"Predicting image {image_path}...")
predict_single_image(model, image_path, categories)
```
